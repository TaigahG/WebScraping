```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import datetime
import time
import smtplib
```

```python
def checking_price():
  import datetime
  URL = 'https://www.amazon.com/s?k=iphone+15+case&ref=nb_sb_noss'
  headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

  page = requests.get(URL, headers=headers)

  soup = BeautifulSoup(page.content, "html.parser")

  Products = soup.select('span.a-text-normal')

  titles = [t.get_text() for t in Products]

  price = soup.select('span.a-offscreen')
  prices = [t.get_text() for t in price]

  rating = soup.select('div.a-size-small')
  ratings = [t.get_text()[0:4] for t in rating]

  votes = [t.get_text()[22:27] for t in soup.select('div.a-size-small')]

  dates = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
  global listItem
  listItem.clear()
  for i in range(len(Products)-7):
      data = {
          "title":titles[i],
          "price":price[i],
          "rating":ratings[i],
          "total_rating":votes[i],
          "date":str(dates)
        }
      listItem.append(data)
```

### Updating it's price every day

```python
while(True):
  checking_price()
  time.sleep(86400)
```

```python
checking_price()
df = pd.DataFrame(listItem, columns = ['title','price','rating','total_rating','date'])
df.head()
```
